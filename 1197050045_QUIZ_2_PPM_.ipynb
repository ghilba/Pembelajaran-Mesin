```python
df['Date'] = df['Date'].fillna(0)
df['AveragePrice'] = df['AveragePrice'].fillna(0)
df['Total Volume'] = df['Total Volume'].fillna(0)
df['Total Bags'] = df['Total Bags'].fillna(0)
df['Small Bags'] = df['Small Bags'].fillna(0)
df['Large Bags'] = df['Large Bags'].fillna(0)
df['XLarge Bags'] = df['XLarge Bags'].fillna(0)
df['type'] = df['type'].fillna(0)
df['year'] = df['year'].fillna(0)
```

```python
df.corr()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

korelasi = df.corr()
plt.figure(figsize=(15, 10))
sns.heatmap(korelasi[(korelasi >= 0.5) | (korelasi <= -0.4)], cmap='viridis', vmax=1.0, vmin=-1.0, linewidths=0.1, 
            annot=True, annot_kws={"size": 8}, square=True).set_title('Korelasi Antar Variabel')
plt.show()
```

```python
avocado_total = df.groupby("region")["Total Volume"].sum().sort_values()
avocado_total.plot(kind="barh", fontsize=10, figsize=(8,40), title='Total Volume Avocado')
plt.show()

bags_total = df.groupby("region")["Total Bags"].sum().sort_values()
bags_total.plot(kind="barh", fontsize=10, figsize=(8,40), title='Total BAgs of Avocado')
plt.show()
```

```python
sns.boxplot(df['Total Bags'])
```

```python
# Scatter plot
fig, ax = plt.subplots(figsize = (20,20))
ax.scatter(df['Total Volume'], df['Total Bags'])

# x-axis label
ax.set_xlabel('Total Volume Avocado')
 
# y-axis label
ax.set_ylabel('Total Bags')
plt.show()
```
